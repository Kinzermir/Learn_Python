```python
def fn_obtained(s1: int, s2: int, s3: int, s4: int, s5: int)->int:
    return s1 + s2 + s3 + s4 + s5


df1['obtained'] = df1[["s1", "s2", "s3", "s4", "s5"]].apply(lambda x:fn_obtained(*x), axis=1)
df1
#df1[["s1", "s2", "s3", "s4", "s5"]]: This selects the columns s1, s2, s3, s4, and s5 from the DataFrame df1.
# For each row x, unpack the values (i.e., pass the values of s1, s2, s3, s4, s5 as arguments) and apply the function fn_obtained. axis=1: This specifies that the function should be applied row-wise.
# Purpose of the code to adds a new column, where each value in this new column is the sum of the corresponding row's 
# Simply multiple inputs and single output.
```

```python
# multiple inputs and multiple outputs.
def fn_obtained(s1: int, s2: int, s3: int, s4: int, s5: int)->tuple[int, int,float, str]: # float islie dala ke ans . points me bh asakta hai.
    total: int = 500
    obtained :int = s1 + s2 + s3 + s4 + s5
    per : float = obtained / total * 100
    grade : str = ""
    if per >= 80:
        grade =  "A+"
    elif per >= 70:
        grade = "A"
    elif per >= 60:
        grade = "B"
    elif per >= 50:
        grade = "C"
    elif per >= 40:
        grade = "D"
    elif per >= 33:
        grade = "E"
    else:
        grade = "F"
    return total, obtained, per, grade


df1[["Total",'obtained', 'percentage', "Grade"]] = df1[["s1", "s2", "s3", "s4", "s5"]].apply(lambda x:fn_obtained(*x), 
                                                                                             axis=1,
                                                                                             result_type='expand') # we add this line for multiple multiple outputs we have multiple inputs.
df1
```

### Map function for remarks

```python
from typing import Dict


remark : Dict[str,str] = {
    "A+" : "Excellent",
    "A" : "Very Good",
    "B" : "Good",
    "C" : "Fair",
    "D" : "Poor",
    "E" : "Very Poor",
    "F": "Fail"
}
# here we applied map function on df1 
#   Column name                 variable
df1['Remark'] = df1.Grade.map(remark)
df1
```
